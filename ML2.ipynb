```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = 'https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv'
df = pd.read_csv(url)
print(df.head())
print(df.info())
print(df.describe(include='all'))
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.subplot(2, 2, 2)
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution')
plt.subplot(2, 2, 3)
sns.countplot(x='Pclass', data=df)
plt.title('Passenger Class Distribution')
plt.subplot(2, 2, 4)
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.tight_layout()
plt.show()
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age by Passenger Class')
plt.subplot(2, 2, 2)
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age by Survival Status')
plt.subplot(2, 2, 3)
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.subplot(2, 2, 4)
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.tight_layout()
plt.show()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.subplot(1, 2, 2)
sns.pairplot(df[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', diag_kind='kde')
plt.title('Pairplot of Age, Fare, Pclass, and Survival')
plt.tight_layout()
plt.show()
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 120], labels=['Child', 'Teen', 'Young Adult', 'Adult', 'Senior'])
plt.figure(figsize=(10, 6))
sns.countplot(x='AgeGroup', hue='Survived', data=df)
plt.title('Survival by Age Group')
plt.show()
df['FareGroup'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])
plt.figure(figsize=(10, 6))
sns.countplot(x='FareGroup', hue='Survived', data=df)
plt.title('Survival by Fare Group')
plt.show()
```
